```python
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from azure.cognitiveservices.vision.computervision.models import OperationStatusCodes
from azure.cognitiveservices.vision.computervision.models import VisualFeatureTypes
from msrest.authentication import CognitiveServicesCredentials

from array import array
import os
from PIL import Image
import sys
import time
```

```python
subscription_key = "86e517e0f72a4d0f9761730dd0baf702"
endpoint = "https://tanya.cognitiveservices.azure.com/"

computervision_client = ComputerVisionClient(endpoint, CognitiveServicesCredentials(subscription_key))
```

```python
#AADHAR CARD
read_image_url = "https://raw.githubusercontent.com/tanya4067/State-Bank-of-India-OCR/main/dwRqcW6.jpg"
read_response = computervision_client.read(read_image_url,  raw=True)
read_operation_location = read_response.headers["Operation-Location"]
operation_id = read_operation_location.split("/")[-1]
while True:
    read_result = computervision_client.get_read_result(operation_id)
    if read_result.status not in ['notStarted', 'running']:
        break
    time.sleep(1)
if read_result.status == OperationStatusCodes.succeeded:
    for text_result in read_result.analyze_result.read_results:
        for line in text_result.lines:
            print(line.text)
            "print(line.bounding_box)"
            

print()
print("Photograph analysis")
tags_result_remote = computervision_client.tag_image(read_image_url )
print("Tags in the remote image: ")
if (len(tags_result_remote.tags) == 0):
    print("No tags detected.")
else:
    for tag in tags_result_remote.tags:
        print("'{}' with confidence {:.2f}%".format(tag.name, tag.confidence * 100))
print()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

file = "sbi-statement-jan13-to-jun13-1-638.webp"
```

```python
im1 = cv2.imread(file, 0)
im = cv2.imread(file)
```

```python
ret,thresh_value = cv2.threshold(im1,180,255,cv2.THRESH_BINARY_INV)

kernel = np.ones((15,15),np.uint8)
dilated_value = cv2.dilate(thresh_value,kernel,iterations = 1)

contours, hierarchy = cv2.findContours(dilated_value,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cordinates = []
```

```python
for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    cordinates.append((x,y,w,h))
    #bounding the images
    if y< 50:
        
        cv2.rectangle(im,(x,y),(x+w,y+h),(0,0,255),1)
        
plt.imshow(im)
```
